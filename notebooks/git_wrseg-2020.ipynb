```
# CONFIGURAÇÕES INICIAIS DE SSH PARA REPOSITÓRIOS PRÓPRIOS DO GITHUB 
# Lembrando: esses comandos funcionarão no seu próprio repositório
# Importante: Executar um comando de cada vez

# Carrega os dados do Google Drive
from google.colab import drive, output
drive.mount('/content/drive')
%cd /content/drive/My\ Drive
# Limpa as saídas do terminal
output.clear()

!ssh-keygen -t rsa -b 4096 -C "seu_email"

# Adicione a saída do comando abaixo no seu perfil do GitHub
!cat /root/.ssh/id_rsa.pub

!ssh-keyscan github.com >> /root/.ssh/known_hosts
!chmod 644 /root/.ssh/known_hosts
!chmod 600 /root/.ssh/id_rsa

!ssh -T git@github.com

!git config --global user.email "seu_email"
!git config --global user.name "seu_username"

%cd /content/drive/My\ Drive/
%ls /content/drive/My\ Drive/

!git clone seu_repositorio.git

%cd seu_repositorio/
%ls /content/drive/My\ Drive/seu_repositorio/
```
